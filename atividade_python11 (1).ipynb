```python
# Resposta questão 2:  
tempo_reposta = round(df['Tempo Resposta'].mean())
df['Tempo Resposta'].fillna(tempo_reposta, inplace= True)
print(df['Tempo Resposta'].mean())
print(df['Tempo Resposta'].max())
print(df['Tempo Resposta'].min())
```

```python
# Resposta questão 3: 
nota_media = round(df['Nota do Consumidor'].mean())
df[ 'Avaliação Reclamação'].fillna(nota_media,inplace= True)
print(df['Nota do Consumidor'].mean())
print(df['Nota do Consumidor'].max())
print(df['Nota do Consumidor'].min())
```

```python
# Resposta questão 4: 
for i in range(0,14,1):
    a = df.loc[df['Tempo Resposta'] == i,:]
    
    print("tempo Resposta , ", i)
    print("Nota do Consumidor", a['Nota do Consumidor'].mean())
    print(' ')
# Aqui podemos ver que tem sim ligação o tempo resposta com a nota do consumidor pois a medidar que a tempo resposta aumenta
# aumenta também a media da nota do consumidor 
```

```python
# Resposta questão 6b: 
Operadoras = df.loc[df['Segmento de Mercado']=='Operadoras de Telecomunicações (Telefonia, Internet, TV por assinatura)']

Bancos = df.loc[df['Segmento de Mercado']=='Bancos, Financeiras e Administradoras de Cartão']

Comércio_Eletrônico = df.loc[df['Segmento de Mercado']=='Comércio Eletrônico']

Fabricantes =  df.loc[df['Segmento de Mercado']=='Fabricantes - Eletroeletrônicos,  Produtos de Telefonia e Informática']

Varejo = df.loc[df['Segmento de Mercado']== 'Varejo']
```

```python
O = Operadoras.loc[Operadoras['Nota do Consumidor'] <= 3]
B = Bancos.loc[Bancos['Nota do Consumidor'] <= 3]
C = Comércio_Eletrônico.loc[Comércio_Eletrônico['Nota do Consumidor'] <= 3]
F = Fabricantes.loc[Fabricantes['Nota do Consumidor'] <= 3]
V = Varejo.loc[Varejo['Nota do Consumidor'] <= 3]
```

```python
o =round(O['Nota do Consumidor'].shape[0] / Operadoras['Nota do Consumidor'].shape[0] * 100)
b =round(B['Nota do Consumidor'].shape[0] / Bancos['Nota do Consumidor'].shape[0] * 100)
c = round(C['Nota do Consumidor'].shape[0] / Comércio_Eletrônico['Nota do Consumidor'].shape[0] * 100)
f = round(F['Nota do Consumidor'].shape[0] / Fabricantes['Nota do Consumidor'].shape[0] * 100)
v = round(V['Nota do Consumidor'].shape[0] / Varejo['Nota do Consumidor'].shape[0] * 100)

lista = [o ,b,c,f,v]
print(lista)
```

```python
lista1 = [Operadoras,Bancos,Comércio_Eletrônico,Fabricantes,Varejo]
valores1 =[]
for l in lista1:
     valores1.append(len(l['Segmento de Mercado']))
```

```python
valores2=[]
for j in range(len(valores1)):
    valores2.append(valores1[j] / 100 * lista[j])
```

```python
print(lista)
```

```python
categoria = ["Operadoras: 36%","Bancos: 31%","Comércio_Eletrônico: 42%","Fabricantes: 32%","Varejo: 24%"]
plt.figure(figsize=(12,4))
plt.bar(categoria,valores2)
plt.bar(categoria,valores1,bottom=valores2 )
plt.show()
```

Eu queria ver dos 5 primeiro segmento de marcado com mais problemais e com as notas mais baixar 

As notas eu coloquei como parametro de ruim igual ou menor que 3 

Em questão de porcentegem comecio eletrônico é que tem mais notas ruim, podemos dizer que é o segmento de mercado com mais problemas de resolver questões de reclamação dos seus clientes
